# Monte Carlo Simulaciones: Decaimiento Radiactivo

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
import time
import pandas as pd
```

```python
def graficar(x_values, y_values):
    plt.plot(x_values, y_values, label='N(t)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Decaimiento Radioactivo")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
nInicial = 10000
lb = .0001
HistN = []
Tiempos = []
t = 0

HistN.append(nInicial)
Tiempos.append(t)

while(nInicial>0):
    nFinal = nInicial
    for i in range(0,nInicial):
        r = random.random()
        if(r < lb):
            nFinal -= 1
    nInicial=nFinal
    t += 1
    HistN.append(nInicial)
    Tiempos.append(t)  
```

```python
print(HistN)
```

```python
HistNArray = np.array(HistN)
tiempoArray = np.array(Tiempos)

graficar(tiempoArray,HistNArray)
```

```python
df = pd.DataFrame({'t':tiempoArray, 'N':HistNArray})
excel_file_path = 'Decaimiento Radiactivo Ld 0001 de 10000.xlsx'
df.to_excel(excel_file_path, index=False)
```

```python
## Experimento de ver cuantas iteraciones le toma en logrlo 
```

```python
Iteraciones = np.zeros(100)
for k in range(0,100):
    nInicial = 10000
    lb = .0001
    HistN = []
    Tiempos = []
    t = 0

    HistN.append(nInicial)
    Tiempos.append(t)

    while(nInicial>0):
        nFinal = nInicial
        for i in range(0,nInicial):
            r = random.random()
            if(r < lb):
                nFinal -= 1
        nInicial=nFinal
        t += 1
        HistN.append(nInicial)
        Tiempos.append(t)
    Iteraciones[k] = len(HistN)
print(Iteraciones)
```

```python
np.mean(Iteraciones)  #Curioso que se tarda casi 10 veces la cantidad inicial si: cant*prob = 1
```

```python
np.std(Iteraciones)
```
