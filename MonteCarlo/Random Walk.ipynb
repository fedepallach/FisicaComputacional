# Monte Carlo Simulaciones: Random Walk

## 2D

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random as rd
import pandas as pd
import time
```

```python
def graficar(x_values, y_values):
    plt.plot(x_values, y_values, label='DistanciaPromedio')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Random Walk 2D")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
pasos = 1000
HistPuntos = np.zeros((1000,2))
HistDists= np.zeros((1000,pasos))
posicion = np.zeros((2))
HistRandoms = np.zeros((1000,pasos))

for i in range (0,1000):
    dist = 0
    posicion[0]=0
    posicion[1]=0
    
    for j in range (1,pasos+1):
        r = rd.randint(1,4)
        HistRandoms[i,j-1] = r
        if r == 1:
            posicion[0] += 1
        if r == 2:
            posicion[1] += 1
        if r == 3:
            posicion[0] -= 1
        if r == 4:
            posicion[1] -= 1
        dist = math.sqrt(math.pow(posicion[0],2)+math.pow(posicion[1],2))
        HistDists[i,j-1] = dist
        
    print(dist) #Distancia final, igual en HistDists[i,pasos-1]
    HistPuntos[i,0] = posicion[0]
    HistPuntos[i,1] = posicion[1]
    
```

```python
DistProms = np.zeros(1000)
for i in range (0,1000):
    DistProms[i] = np.mean(HistDists[:,i])
    #print(DistProms[i])

graficar(range(1,1001),DistProms)
```

### A Excel

```python
df = pd.DataFrame(DistProms, columns=['DistanciaPromedio'])
excel_file_path = 'CaminarAleatorio2D.xlsx'
df.to_excel(excel_file_path, index=False)
```

```python
DistProms = np.zeros(1000)
for i in range (0,1000):
    DistProms[i] = np.mean(np.abs(HistDists[:,i]))
    #print(DistProms[i])

graficar(range(1,1001),DistProms)
```

### A Excel

```python
df = pd.DataFrame(DistProms, columns=['DistanciaPromedio'])
excel_file_path = 'CaminarAleatorio1D.xlsx'
df.to_excel(excel_file_path, index=False)
```

## 2D con direccion de paso variable

```python
pasos = 1000
HistPuntos = np.zeros((1000,2))
HistDists= np.zeros((1000,pasos))
posicion = np.zeros((2))
HistRandoms = np.zeros((1000,pasos))

for i in range (0,1000):
    dist = 0
    posicion[0]=0
    posicion[1]=0
    rd.seed(time.time())
    for j in range (1,pasos+1):
        r0 = rd.uniform(-1,1)
        r1 = rd.uniform(-1,1)
        HistRandoms[i,j-1] = r0 #faltaria poner r1 igual...
        posicion[0] += r0
        posicion[1] += r1
        
        dist = math.sqrt(math.pow(posicion[0],2)+math.pow(posicion[1],2))
        HistDists[i,j-1] = dist
        
    print(dist) #Distancia final, igual en HistDists[i,pasos-1]
    HistPuntos[i,0] = posicion[0]
    HistPuntos[i,1] = posicion[1]
    
```

```python
DistProms = np.zeros(1000)
for i in range (0,1000):
    DistProms[i] = np.mean(HistDists[:,i])
    #print(DistProms[i])

graficar(range(1,1001),DistProms)
```

### A Excel

```python
df = pd.DataFrame(DistProms, columns=['DistanciaPromedio'])
excel_file_path = 'CaminarAleatorio2D.xlsx'
df.to_excel(excel_file_path, index=False)
```
