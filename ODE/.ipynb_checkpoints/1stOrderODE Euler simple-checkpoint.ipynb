# Euler simple

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Euler Simple Funcion

```python
  """
    Solves the ODE y' = f(x, y) using Euler's method.

    Parameters:
    f: Function that defines the ODE (dy/dx = f(x, y))
    x0: Initial value of x
    y0: Initial value of y
    h: Step size
    num_steps: Number of steps to take

    Returns:
    x_values: Array of x values
    y_values: Array of y values
    """
def euler_method(f, x0, y0, h, num_steps):
  
    x_values = np.zeros(num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    x_values[0] = x0
    y_values[0] = y0

    for i in range(1, num_steps + 1):
        y_values[i] = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])
        x_values[i] = x_values[i-1] + h

    return x_values, y_values
```

## Euler simple aplicacion

```python

# Example usage
if __name__ == "__main__":
    k = .2
    # Define the ODE y' = f(x, y) = -2 * x * y
    def f(x, y):
        return -k * y

    # Initial conditions
    x0 = 0
    y0 = 10

    # Step size
    h = .1

    # Number of steps
    num_steps = 200

    # Solve the ODE using Euler's method
    x_values, y_values = euler_method(f, x0, y0, h, num_steps)

    # Plot the solution
    graficar(x_values, y_values)
```

```python
def graficar(x_values, y_values):
    plt.plot(x_values, y_values, label='Euler Method')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Solving ODE using Euler's Method")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Grafica del error en func de x

```python
meanAbsoluteErrorList = []
```

```python
def valorDeFuncion(x,k):
    valor = 10*math.exp(-k*x)
    return valor
```

```python
#Grafica del error en func de x
k=.2
listRealValues = []
listErrors = []
for i in range (0,num_steps+1):
    #print(x_values[i])
    #print(y_values[i])
    valorReal = valorDeFuncion(x_values[i],k)
    listRealValues.append(valorReal)
    listErrors.append(valorReal-y_values[i])
#print(listRealValues)
arrayRealValues = np.array(listRealValues)
#print(arrayRealValues)
arrayErrorValues = np.array(listErrors)
#print(arrayErrorValues)

# Plot the solution
plt.plot(x_values, arrayErrorValues, label='Euler Method Error')
plt.xlabel('x')
plt.ylabel('Error')
plt.title("Solving ODE using Euler's Method")
plt.legend()
plt.grid(True)
plt.show()

#Mean Absolute Error
meanAbsoluteError = np.mean(np.abs(arrayErrorValues))
meanAbsoluteErrorList.append(meanAbsoluteError)

```
