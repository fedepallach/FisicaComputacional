# Euler modificado y mejorado

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Graficar

```python
def graficar(x_values, y_values):
    plt.plot(x_values, y_values,label='Euler modificado')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Solving ODE using Euler's Method")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Euler simple

```python
def euler_simp_method(f, xi, yi, h):
    y = yi + h * f(xi, yi)
    return y
```

## Euler modificado

```python
def euler_modif_method(f, x0, y0, h, num_steps):
  
    x_values = np.zeros(num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    ymit = 0
    
    x_values[0] = x0
    y_values[0] = y0
    
    for i in range(1, num_steps + 1):
        ymit = euler_simp_method(f,x_values[i-1],y_values[i-1],h/2)
        pendienteMitad = f((x_values[i-1]+h/2), ymit)
        y_values[i] = y_values[i-1] + h * pendienteMitad
        x_values[i] = x_values[i-1] + h

    return x_values, y_values
```

## Euler modificado (Main)

```python
#Valores - Parametros - Condiciones
x0 = 0
y0 = 10
k = .2   #func de superficie de contacto, litros en la alberca, prop del material, interac igual si viento otra cosa dif no?
h = .1
num_steps = 200

#Funcion
def f(x, y):
    return -k * y

# Resolver
x_values, y_values = euler_modif_method(f, x0, y0, h, num_steps)

# Plot the solution
#label = 'ODE Euler Modificado'
graficar(x_values, y_values)
```

## Grafica del error en func de x (modif)

```python
meanAbsoluteErrorList = []   #solo correrlo al inicio
```

```python
def valorDeFuncion(x,k):
    valor = 10*math.exp(-k*x)
    return valor
```

```python
#Grafica del error en func de x
k =.2
listRealValues = []
listErrors = []
for i in range (0,num_steps+1):
    #print(x_values[i])
    #print(y_values[i])
    valorReal = valorDeFuncion(x_values[i],k)
    listRealValues.append(valorReal)
    listErrors.append(valorReal-y_values[i])
#print(listRealValues)
arrayRealValues = np.array(listRealValues)
#print(arrayRealValues)
arrayErrorValues = np.array(listErrors)
#print(arrayErrorValues)

# Plot the solution
plt.plot(x_values, arrayErrorValues, label='Euler Mod Method Error')
plt.xlabel('x')
plt.ylabel('Error')
plt.title("Solving ODE using Euler's Method")
plt.legend()
plt.grid(True)
plt.show()

#Mean Absolute Error
meanAbsoluteError = np.mean(np.abs(arrayErrorValues))
meanAbsoluteErrorList.append(meanAbsoluteError)

```

## Euler Mejorado

```python
def euler_mej_method(f,x0,y0,h,n):
    x_values = np.zeros(n+1);
    y_values = np.zeros(n+1);
    
    x_values[0]=x0
    y_values[0]=y0
    
    for i in range(1,n+1):
        #x_values
        x_values[i] = x_values[i-1] + h
        
        #y_values
        pend1 = f(x_values[i-1],y_values[i-1])
        ysig = euler_simp_method(f,x_values[i-1],y_values[i-1],h)
        pend2 = f(x_values[i],ysig)
        y_values[i] = y_values[i-1] + (h/2)*(pend1+pend2) 
        
    return x_values, y_values
```

## Euler Mejorado (Main)

```python
#Valores - Parametros - Condiciones
x0 = 0
y0 = 10
k = .2   #func de superficie de contacto, litros en la alberca, prop del material, interac igual si viento otra cosa dif no?
h = .1
num_steps = 200

#Funcion
def f(x, y):
    return -k * y

# Resolver
x_values, y_values = euler_mej_method(f, x0, y0, h, num_steps)

# Plot the solution
#label = 'ODE Euler Modificado'
graficar(x_values, y_values)
```
