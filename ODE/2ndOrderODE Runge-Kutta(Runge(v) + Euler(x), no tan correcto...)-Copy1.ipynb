# Euler modificado y mejorado

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Graficar

```python
def graficar(x_values, y_values):
    plt.plot(x_values, y_values,label='Euler modificado')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Solving ODE using Euler's Method")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Euler simple

```python
def euler_simp_p(f, xi, yi, ypi):
    p = f(xi, yi, ypi)
    return p
```

## Euler modificado

```python
def euler_modif_p(f, x, y, yp, h):
    ymit = y + euler_simp_p(f,x,y) * h/2
    pendienteMitad = f(x+(h/2), ymit) ###problema!!! porque no esatas mandando la xmit igual! que aparece en la f2()! (porque en SHM depende de x (en este caso y para obtener v (en este caso y no?)))
    return pendienteMitad
```

## Euler modificado para k2

```python
def euler_modif_p_k2(f, x, y, yp, h):
    ymit2 = y + euler_modif_p(f,x,y,h) * h/2  # igual es: y + k1  * h/2
    pendienteMitad2 = f(x+(h/2), ymit2)
    return pendienteMitad2
```

## Extra para k3

```python
def para_k3(f,x,y,yp,h):
    pend = f(x + h/2 , y + euler_modif_p_k2(f,x,y,h) * h)
    return pend
```

## Euler Mejorado (no se usa)

```python
def euler_mej_p(f,x,y,h):
    pend1 = f(x,y)
    ysig = y + pend1 * h
    pend2 = f(x+h,ysig)
    pendProm = (pend1+pend2)/2
    return pendProm
```

## Runge-Kutta

```python
def runge_kutta(f,x,y,yp,h):   #donde y(x) y yp = dy/dx
    k0 = euler_simp_p(f,x,y,yp)
    k1 = euler_modif_p(f,x,y,yp,h)
    k2 = euler_modif_p_k2(f,x,y,yp,h)
    k3 = para_k3(f,x,y,yp,h)
    resultado = y + (h/6) * (k0 + 2*k1 + 2*k2 + k3)
    return resultado
```
